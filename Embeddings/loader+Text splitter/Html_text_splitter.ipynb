### HTML Text Splitter (see notes)

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string="""
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>My Basic HTML Page</title>
</head>
<body>
    <h1>Welcome to My Page</h1>
    <p>This is a simple HTML page.</p>

    <h2>Sections</h2>
    <p>Here's a list of some sections:</p>

    <ul>
        <li>Introduction</li>
        <li>Main Content</li>
        <li>Conclusion</li>
    </ul>

    <p>This page is located in Shimla, Himachal Pradesh, on March 26, 2025.</p>
    <p>Enjoy your visit!</p>
</body>
</html>
"""
header_to_split_on=[
    ("h1","Header_1"),
    ("h2","Header_2")
    ]

html_splitter=HTMLHeaderTextSplitter(header_to_split_on)
html_header_splits=html_splitter.split_text(html_string)
html_header_splits
```
