```python
##web based loader
from langchain_community.document_loaders import WebBaseLoader
loader_2=WebBaseLoader(web_path=("https://lilianweng.github.io/posts/2023-06-23-agent/"))
loader_2
```

```python
loader_2.load()#full content 
```

```python
##web based loader with customize data
from langchain_community.document_loaders import WebBaseLoader
import bs4
loader_3=WebBaseLoader(web_path=("https://lilianweng.github.io/posts/2023-06-23-agent/",),
                       bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                       class_=("post-title","post-content","post-header")
                       ))
                       )
loader_3
```

```python
loader_3.load()#ti 
```
